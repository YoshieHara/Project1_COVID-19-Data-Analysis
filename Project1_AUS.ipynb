```python
Question:

1. The trend of COVID-19 infection cases in Australia statewide for the first week of January 2022ï¼Ÿ
2. Which states in Australia has the highest confirmed casess for the first week of January 2022?
3. Which states in Australia has the highest death casess for the first week of January 2022?
```

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Define the API endpoint URL
#url = "https://covid-api.com/api/reports?date=2023-02-09"
#url = "https://covid-api.com/api/reports?iso=AUS"
url1 = "https://covid-api.com/api/reports?date=2022-01-01&iso=AUS"
url2 = "https://covid-api.com/api/reports?date=2022-01-02&iso=AUS"
url3 = "https://covid-api.com/api/reports?date=2022-01-03&iso=AUS"
url4 = "https://covid-api.com/api/reports?date=2022-01-04&iso=AUS"
url5 = "https://covid-api.com/api/reports?date=2022-01-05&iso=AUS"
url6 = "https://covid-api.com/api/reports?date=2022-01-06&iso=AUS"
url7 = "https://covid-api.com/api/reports?date=2022-01-07&iso=AUS"

# Make a GET request to the API
response1 = requests.get(url1)
response2 = requests.get(url2)
response3 = requests.get(url3)
response4 = requests.get(url4)
response5 = requests.get(url5)
response6 = requests.get(url6)
response7 = requests.get(url7)

# Parse the JSON response into a Python dictionary
data1 = response1.json()
data2 = response2.json()
data3 = response3.json()
data4 = response4.json()
data5 = response5.json()
data6 = response6.json()
data7 = response7.json()

# pd.set_option("display.max_colwidth", None)  
# df1

# Convert the dictionary to a Pandas DataFrame
df1 = pd.json_normalize(data1['data']) 
df2 = pd.json_normalize(data2['data']) 
df3 = pd.json_normalize(data3['data']) 
df4 = pd.json_normalize(data4['data']) 
df5 = pd.json_normalize(data5['data']) 
df6 = pd.json_normalize(data6['data']) 
df7 = pd.json_normalize(data7['data']) 
```

```python
# Combine the DataFrames of Jan - March
df = pd.concat([df1, df2, df3, df4, df5, df6, df7], ignore_index=True)
df
```

```python
# Checking the data 
df["date"].value_counts()
```

```python
# Drop the columns from the dataFrame
df = df.drop(columns=["last_update"])
df = df.drop(columns=["region.iso"])
df = df.drop(columns=["region.name"])
df = df.drop(columns=["region.cities"])
df
```

```python
total_confirmed_per_day = df.groupby('date')['confirmed'].sum()
total_confirmed_per_day
```

```python


# Group by date and sum the confirmed cases for each day
total_confirmed_per_day = df.groupby('date')['confirmed'].sum().reset_index()

# Create a line chart for total confirmed cases per day
plt.plot(total_confirmed_per_day['date'], total_confirmed_per_day['confirmed'], marker='o')
plt.title('Daily Confirmed Cases in Australia 2022/01/01 ~ 2022/01/07')
plt.xlabel('Date')
plt.ylabel('Total Confirmed Cases')
plt.xticks(rotation=45)
plt.show()

print(total_confirmed_per_day)
```

```python


# Plotting the line chart with different colors for each province
fig, ax = plt.subplots(figsize=(10, 6))

for province, province_df in df.groupby('region.province'):
    ax.plot(province_df['date'], province_df['active'], label=province, marker='o')

plt.title('Daily Active Cases by Province')
plt.xlabel('Date')
plt.ylabel('Active Cases')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
# Convert the 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Print actual numbers for each province
for province, province_df in df.groupby('region.province'):
    print(f"\nProvince: {province}")
    print("Date\t\tActive Cases")
    for i, date in enumerate(province_df['date']):
        print(f"{date}\t\t{province_df['active'].iloc[i]}")
```

```python
# Group by 'region.province' and aggregate the sum of other columns
aggregated_df = df.groupby('region.province').agg({
    'confirmed': 'sum',
    'deaths': 'sum',
    'recovered': 'sum',
}).reset_index()

# Display the aggregated DataFrame
aggregated_df
```

```python
plt.figure(figsize=(8, 4))
plt.bar(aggregated_df['region.province'], aggregated_df['confirmed']/1000, width=0.6)
plt.title('Total Confirmed Cases by Provinces')
plt.xlabel('Proviince')
plt.ylabel('Total cases (in thousand)')
plt.xticks(rotation=45)
plt.show()

print(aggregated_df[['region.province', 'confirmed']])
```

```python
plt.figure(figsize=(8, 4))
plt.bar(aggregated_df['region.province'], aggregated_df['deaths'], width=0.6)
plt.title('Total Death Cases by Provinces')
plt.xlabel('Proviince')
plt.ylabel('Total Death Cases')
plt.xticks(rotation=45)
plt.show()
print(aggregated_df[['region.province', 'deaths']])
```
